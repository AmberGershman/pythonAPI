### Humidity Heatmap
* Configure gmaps.

### Create new DataFrame fitting weather criteria

```python
ideal_conditions = weather_py.loc[weather_py["Max Temperature"] < 90, :]
ideal_conditions = weather_py.loc[weather_py["Max Temperature"] > 45, :]
ideal_conditions = weather_py.loc[weather_py["Humidity"] < 60, :]
ideal_conditions.dropna(axis = 1)
ideal_conditions
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = ideal_conditions
hotel_df["Hotel Name"] = ""
hotel_df["Hotel Rating"] = ""
hotel_df["Hotel Address"] = ""
warnings.filterwarnings("ignore")
hotel_df
```

```python
warnings.filterwarnings("ignore")
hotel_name = []
hotel_rating = []
vicinity = []
params = {
    "radius": 50000,
    "types": "hotel",
    "keyword": "hotel",
    "key": g_key
}

for index, row in hotel_df.iterrows():
    lat = row["Latitude"]
    lng = row["Longitude"]
    
    params["location"] = f"{lat},{lng}"
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    name_address = requests.get(base_url, params=params)
    name_address = name_address.json()

    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["results"][0]["name"]
        hotel_df.loc[index, "Hotel Address"] = name_address["results"][0]["vicinity"]
        hotel_df.loc[index, "Hotel Rating"] = name_address["results"][0]["rating"]
    except (KeyError, IndexError):
        print("Nope - can't find data for this one. You DON'T GET TO VACATION HERE")
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row


# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map

gmaps.configure(api_key=g_key)
locations = hotel_df[["Latitude", "Longitude"]]
humid = hotel_df["Humidity"]

# Display figure
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humid, 
                                 dissipating=False, max_intensity=10,
                                 point_radius = 1)

hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content= hotel_info
)


fig.add_layer(heat_layer)
fig.add_layer(hotel_layer)
fig
```
